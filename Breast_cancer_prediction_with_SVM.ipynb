# Now Using GridSerachCV(hyperperametar Tuning)

```python
cancer_data = pd.read_csv('Breast_Cancer_Diagnostic.csv')
cancer_data
```

```python
x = cancer_data.drop(['diagnosis'],axis=1)
x
```

```python
y=cancer_data['diagnosis']
y
```

```python
type(y)
```

### Now scaling features 

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
x_scaler = mms.fit_transform(x)
x_scaler
```

### using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

clf = GridSearchCV((SVC()),{
    'gamma':['auto','scale'],
    'C':[1,5,10,15],
    'kernel':['rbf','linear','poly']
},cv=5)
clf.fit(x_scaler,y)
clf.cv_results_
```

```python
data = pd.DataFrame(clf.cv_results_)
data
```

```python
best_results = clf.best_score_
best_results
```

```python
best_params = clf.best_params_
best_params
```

## Cross Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
sv1 = cross_val_score(SVC(kernel='poly',C=10,gamma = 'scale'),x_scaler,y,cv=5)
sv1
```

```python
sv1.mean()
```

```python
sv2 = cross_val_score(SVC(kernel='rbf',C=1,gamma = 'auto'),x_scaler,y,cv=5)
sv2
```

```python
sv2.mean()
```
